```python
from sklearn.linear_model import LinearRegression
import pandas as pd
```

```python
df=pd.read_csv("audi.csv")
```

```python
df=df.drop(columns=["index","href","MileageRank","PriceRank","PPYRank","Score"])
```

```python
df.head(3)
```

```python
df.info()
```

```python
df.columns=['il','kasa','mil','motor','ps','vites','yanacaq','sahiblerin_sayı','qiymet','ppy']
```

```python
df.head(3)
```

```python
df['motor']=df['motor'].str.replace("L","")
```

```python
df.head(3)
```

```python
df['motor']=pd.to_numeric(df["motor"])
```

```python
df.head(3)
```

```python
df = pd.get_dummies(df, columns=["kasa", "vites", "yanacaq"], drop_first=True)
df = df.astype(int)
```

```python
df.head(3)
```

```python
y=df[["qiymet"]]
x=df.drop("qiymet",axis=1)
```

```python
lm = LinearRegression()
model=lm.fit(x, y)
```

```python
model
```

```python
model.coef_
```

```python
model.score(x,y)
```
